### Q3. What is a box tuple, and how does it work?

### Answer: Any image pixels can be represented as x-coordinates and y-coordinates. 
### Pillow functions and methods are taking a tuple of 4 integer coordinates and that integer coordinates represented as rectangular region of an image. 
### The 4 integers are:
### 1. Left: x-coordinate of the left edge of the box.
### 2. Top: y-coordinate of the top edge of the box.
### 3. Right: x-coordinate of the right edge of the box.
### 4. Bottom: y-coordinate of the bottom edge of the box.

### Q5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?

### Answer:

```python
from PIL import Image
i = Image.new('RGBA', (100, 100), 'red')
i.save('Image.png')
i
```

### Q7. What module contains Pillow’s shape-drawing code?

### Answer: ImageDraw.Draw() method.

```python
from PIL import Image, ImageDraw
i = Image.new('RGBA', (100, 100), 'red')
i.save('Image.png')
d=ImageDraw.Draw(i)
```
